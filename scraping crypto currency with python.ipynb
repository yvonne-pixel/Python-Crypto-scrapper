```python
soup =  BeautifulSoup(response.content,'html.parser')
```

put everything together inside for loop

```python
name = []
price = []
change_24h = []
change_1day = []
change_7days = []
market_cap = []
volume_24h = []

for result in results:
    try:
        name.append(result.find('p',{'class':'sc-1eb5slv-0 iworPT'}).get_text())
        price.append(result.find('div', attrs={'class':'sc-131di3y-0'}).text)
        changes = result.find_all('span',{'class':'sc-15yy2pl-0 kAXKAX'})
        change_24h.append(changes[0].text)
        change_1day.append(changes[1].text)
        change_7days.append(changes[2].text)
        market_cap.append(result.find('span',{'class':'sc-1ow4cwt-1 ieFnWP'}).text)
        volume_24h.append(result.find('p',{'class':'sc-1eb5slv-0 hykWbK'}).text)
    except:
        name.append('N/A')
    
```
